## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Convert data types if needed



## Handling Missing Data

```python
# Check Missing values

df.isnull().sum()
```

```python
# Handle Missing Values – Numerical Columns
df['Decibel_Level'].fillna(df['Decibel_Level'].mean())
```

```python

df['Cost_of_Living_Index'].fillna(df['Cost_of_Living_Index'].median())
```

```python
# Handle Missing Values – Categorical Columns
df['City'].fillna(df['City'].mode()[0])
```

```python
# Drop Rows
df.dropna(thresh=len(df.columns)*0.7, inplace=True)
```

## Handling Duplicate Records

```python
# Check Duplicates
df.duplicated().sum()
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

## Data Formatting & Standardization

```python
# Rename Columns (Clean Naming)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# Convert Data Types

df['year'] = df['year'].astype(int)
```

```python
# Fix Text Formatting
df['city'] = df['city'].str.title()
```

## Derived Features (Feature Engineering)

```python
# Create New Column from Existing Data
df['happiness_category'] = np.where(df['happiness_score'] >= 7, 'High', 'Low')
```

## Data Transformation

```python
# Scaling (Normalization)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['happiness_score_scaled']] = scaler.fit_transform(df[['happiness_score']])
```

```python
#Encoding Categorical Variables
# Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['city_encoded'] = le.fit_transform(df['city'])
```

```python
# Final Data Quality Check
df.isnull().sum()
df.describe()
df.info()
```

## Stage 3 – EDA and Visualizations



## UNIVARIATE ANALYSIS

### Countplot – Cities by Country

### Countplot equivalent using bar chart - Traffic Density

```python
plt.figure(figsize=(10,6))
traffic_counts = df['traffic_density'].value_counts()
plt.figure()
traffic_counts.plot(kind='bar')
plt.title("Traffic Density Distribution")
plt.xlabel("Traffic Density")
plt.ylabel("Number of Cities")
plt.show()
```

Majority cities fall under Low & Medium traffic

Very High traffic cities are fewer

Traffic congestion is a key urban planning issue, not universal

### Histogram - Happiness Score

```python
plt.figure()
plt.hist(df['happiness_score'], bins=8)
plt.title("Distribution of Happiness Score")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()
```

Happiness scores are mostly mid-range

Very low happiness cities exist but are limited

Improvement programs should target average-performing cities

### Boxplot - Cost of Living Index

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['cost_of_living_index'])
plt.title("Cost of Living Index Distribution")
plt.ylabel("Cost of Living Index")
plt.show()
```

Cost of living varies widely

Presence of outliers → expensive cities

Median cost is more reliable than mean for decisions

### BIVARIATE ANALYSIS

### Green Space vs Happiness – Scatter Plot

```python
plt.figure()
plt.scatter(df['green_space_area'], df['happiness_score'])
plt.xlabel("Green Space Area")
plt.ylabel("Happiness Score")
plt.title("Green Space vs Happiness")
plt.show()
```

Cities with more green space tend to be happier

Strong negative trend if green space reduces

Urban greenery directly impacts mental well-being

### Traffic Density vs Avg Happiness – Bar Chart

```python
avg_happiness = df.groupby('traffic_density')['happiness_score'].mean()

plt.figure()
avg_happiness.plot(kind='bar')
plt.xlabel("Traffic Density")
plt.ylabel("Average Happiness Score")
plt.title("Traffic Density vs Average Happiness")
plt.show()
```

Low traffic cities show higher happiness

Medium & High traffic cities show reduced happiness

Traffic control = quality of life improvement lever

### Correlation Heatmap

```python
plt.figure(figsize=(15,10))
corr = df[['decibel_level',
           'green_space_area',
           'air_quality_index',
           'happiness_score',
           'cost_of_living_index',
           'healthcare_index']].corr()

plt.figure()
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```

Happiness is positively linked with:

Green Space

Healthcare Index

Negatively impacted by:

Noise (Decibel)

Poor Air Quality

Happiness is multi-factor driven

### MULTIVARIATE ANALYSIS

```python
grouped = df.groupby(['year', 'traffic_density'])['happiness_score'].mean().unstack()

plt.figure()
grouped.plot()
plt.xlabel("Year")
plt.ylabel("Average Happiness Score")
plt.title("Happiness Trend by Year & Traffic Density")
plt.show()
```

Across years, Low traffic cities remain happier

Traffic impact is consistent over time

Long-term urban policies matter

### Month × Traffic Density – Pivot Table

```python
pivot = pd.pivot_table(
    df,
    values='happiness_score',
    index='month',
    columns='traffic_density',
    aggfunc='mean'
)

pivot
```

Seasonal impact visible on happiness

Traffic effect varies month-wise

Useful for seasonal city planning & tourism strategy
